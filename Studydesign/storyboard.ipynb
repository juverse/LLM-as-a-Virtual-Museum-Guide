## Storyboard

Storyboard – VR/AR Museumsumgebung mit LLM-Interaktion
Ziel:
Realistische, immersive Museumsumgebung, zunächst in VR, später auch in AR mit mehreren Usern. Die Interaktion erfolgt entweder per Sprache, über Knöpfe oder durch eine Kombination beider Methoden. User behalten jederzeit die Kontrolle über das Erlebnis.

Variante 1: Sprachinteraktion
Beschreibung:
User interagiert ausschließlich per Sprache. Um eine Audioaufnahme zu starten, wird ein Knopf gedrückt.
User kann das LLM jederzeit unterbrechen oder stoppen. Anfragen sind explzit, die Kontrolle liegt beim User.

1. User kommt in den VR Raum
2. Ein Knopf wird gedrückt, um die Sprachaufnahme zu aktivieren. 
3. User sagt Hallo
3. Das LLM begrüßt und stellt sich vor.
4. User kann LLM Fragen stellen und bekommt hilfreiche Antworten
5. User telepotiert durch das Museum
6. User steht vor einem Bild, in einem Kontextbereich. Kontext wird angepasst.
7. User fragt das LLM explizit nach Informationen bzw. nach einer kurzen Einführung.
8. LLM bekommt Kontextbereich und antwortet und gibt weitere Themenvorschläge
9. Beim Weitergehen der Users, verstummt das LLM automatisch.
10. Eine Ansprache des LLMs ist jederzeit möglich, auch außerhalb des Kontextbereichs


Variante 2: Interaktion über Knöpfe
Beschreibung:
Es gibt drei Knöpfe zur Steuerung des LLM:

Erzähle etwas/Option 1, Option 2, Stopp
Das LLM endet immer mit entweder "Soll ich weiterreden?" (Irgendein Button außer Stopp: Weiter, Stopp: Nein) oder "Soll ich mehr zu A oder B erzählen?" enden.
der User kann das llm ständig unterbrechen oder es mit einem Knopf zum stoppen bringen


 
1. User kommt in den VR Raum
2. User drückt Knopf "Erzähle mehr"
3. LLM begrüßt und stellt sich vor
5. User telepotiert durch das Museum
6. User steht vor einem Bild,  in einem Kontextbereich, LLM bekommt den Kontextbereich, redet aber nicht
7. User drückt, "Erzähle mehr" um etwas über das Bild zu erfahren
8. LLM antwortet, in dem Themengebiet oder in einem anderem, je nach Knopfdruck
9. User geht weiter aus dem Kontextbereich heraus, das LLM hört auf zu reden
10. User kann auch außerhalb der Kontextbereiche den Knopf "Erzähle mehr" drücken, das LLM erzählt etwas über das Museum

Variante 3: Kombination aus Sprache und Knöpfen
Beschreibung:
Die Interaktion erfolgt flexibel – entweder per Sprache oder über Knöpfe. Beide Eingabemethoden sind jederzeit verfügbar.
Das LLM endet immer mit einem Themenvorschlag worüber geredet werden könnte. Der kann das LLM ständig unterbrechen oder es mit einem Knopf zum stoppen bringen

1. User kommt in den VR Raum
2. Ein Knopf wird gedrückt, um die Sprachaufnahme zu aktivieren. 
3. User sagt "hallo" alternative drückt er nur den Knop "Erzähle mehr"
4. LLm begrüßt und stellt sich vor
5. User kann per Sprachaufnahme etwas sagen oder den Knopf "erzähle mehr" drücken
6. LLM antwortet entsprechend
7. User telepotiert durch das Museum
8. User steht vor einem Bild, in einem Kontextbereich, LLM bekommt den Kontextbereich, redet aber nicht
9. User drückt den Knopf, "erzähle mehr" oder stellt per Sprachaufnahme eine Frage
10. LLM antwortet 
11. User geht weiter zum nächsten Kontext, das LLM hört auf zu reden, sobald der User nicht mehr im Kontext ist
12. User kann auch außerhalb der Kontextbereiche den Knopf "Erzähle mehr" drücken oder per Sprachaufnahme etwas fragen, das LLM erzählt etwas über das Museum.

Zum Visuellen Feedback:
Der Kontextbereich für ein Bild ist per default grau.
sobald der User in einem Kontextbereich ist, leuchtet der Boden weiß auf. Verlässt er ihn, wird er wieder grau

Variante 2: Interaktion über Knöpfe
Beschreibung:
Es gibt drei Knöpfe zur Steuerung des LLM:

„Erzähle mehr“

„Erzähle etwas anderes“

„Stopp“

Der/die Nutzer:in betritt den VR-Raum.

Ein Knopf wird gedrückt, um die Interaktion zu starten.

Das LLM begrüßt und stellt sich vor.

Das LLM spricht, während der/die Nutzer:in über Knöpfe Feedback geben kann.

Das LLM endet jede Passage mit einem Themenvorschlag zur Weiterführung.

Der/die Nutzer:in teleportiert sich durch das Museum.

Das LLM erkennt den Standort des/der Nutzer:in (Kontextsynchronisation).

Vor einem Kunstwerk drückt der/die Nutzer:in auf „Erzähle mehr“.

Zusätzlich kann der/die Nutzer:in das LLM nach allgemeinen Informationen oder einer Einführung fragen.

Beim Weitergehen stoppt das LLM automatisch.

Eine Unterbrechung ist jederzeit über den „Stopp“-Knopf möglich.

Die Steuerung bleibt durchgehend beim Nutzer / bei der Nutzerin.

Variante 3: Kombination aus Sprache und Knöpfen
Beschreibung:
Die Interaktion erfolgt flexibel – entweder per Sprache oder über Knöpfe. Beide Eingabemethoden sind jederzeit verfügbar.

Der/die Nutzer:in betritt den VR-Raum.

Die Interaktion wird durch Sprache („Hallo“) oder Knopfdruck gestartet.

Das LLM begrüßt und stellt sich vor.

Der/die Nutzer:in kann entweder sprechen oder Knöpfe drücken, um mit dem LLM zu interagieren.

Das LLM endet jeden Abschnitt mit einem Themenvorschlag.

Der/die Nutzer:in teleportiert sich durch das Museum.

Das LLM erkennt den aktuellen Standort des/der Nutzer:in.

Vor einem Kunstwerk erfolgt die Interaktion per Sprachbefehl oder durch Knopfdruck („Erzähle mehr“).

Fragen nach einer Einführung oder zum LLM selbst sind jederzeit möglich.

Beim Standortwechsel hört das LLM automatisch auf zu sprechen.

Eine Unterbrechung ist jederzeit durch Sprache oder den „Stopp“-Knopf möglich.

Der/die Nutzer:in kann das LLM jederzeit erneut ansprechen – die Kontrolle bleibt bei ihm/ihr.
